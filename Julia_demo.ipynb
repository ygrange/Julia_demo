## Plotting

### Gadfly


Gadfly is a native Julia plotting package, based on the Grammar of Graphics syntax, and thus with a similar flavour to R's ggplot package. Looks like the package redefines the division function that was defined by the SIUnits package, so you get a warning. 


```julia
# Pkg.add("Gadfly")
import Gadfly
gf=Gadfly
```

```julia
xs = 1:100
ys = cos(xs)

gf.plot(x=xs, y=ys)
```

### Profiling
I had to show something about profiling of course!

```julia
#Pkg.add("ProfileView")
using ProfileView   # This makes the output a bit more understandable for me...
```

```julia
function profile_test(n)
    for i = 1:n
        A = randn(100,100,20)
        m = maximum(A)
        Afft = fft(A)
        Am = mapslices(sum, A, 2)
        B = A[:,:,5]
        Bsort = mapslices(sort, B, 1)
        b = rand(100)
        C = B.*b
    end
end

profile_test(1)  # run once to trigger compilation
Profile.clear()  # in case we have any previous profiling data
@profile profile_test(10)
```
